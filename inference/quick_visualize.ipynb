```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import random

# Set directories for input and ground truth images
input_dir = "/home/hamzaz/hamza/dl/image-deblurring/inference/dataset/motion/testrealblur/RealBlur-J/input"
target_dir = "/home/hamzaz/hamza/dl/image-deblurring/inference/dataset/motion/testrealblur/RealBlur-J/target"

def display_paired_images(input_dir, target_dir, num_pairs=2):
    # Get sorted lists of files from both directories
    input_images = sorted([
        os.path.join(input_dir, f)
        for f in os.listdir(input_dir)
        if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))
    ])
    
    target_images = sorted([
        os.path.join(target_dir, f)
        for f in os.listdir(target_dir)
        if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))
    ])
    
    # Pair input and target images (assume same file names in both directories)
    paired_images = list(zip(input_images, target_images))
    
    # Randomly select the desired number of pairs
    selected_pairs = random.sample(paired_images, min(len(paired_images), num_pairs))
    
    # Create a grid for the paired images
    fig, axes = plt.subplots(num_pairs, 2, figsize=(10, num_pairs * 3))  # 2 columns, num_pairs rows
    for i, (input_path, target_path) in enumerate(selected_pairs):
        # Load input image
        input_img = mpimg.imread(input_path)
        axes[i, 0].imshow(input_img)
        axes[i, 0].axis('off')
        axes[i, 0].set_title("Input", fontsize=10)
        
        # Load ground truth image
        target_img = mpimg.imread(target_path)
        axes[i, 1].imshow(target_img)
        axes[i, 1].axis('off')
        axes[i, 1].set_title("Ground Truth", fontsize=10)
    
    plt.tight_layout()
    plt.show()

display_paired_images(input_dir, target_dir)
```
