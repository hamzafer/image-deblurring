```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Data for inference times
inference_data = {
    "Model": [
        "Improved Model",
        "Fine-tuned on RealBlurJ",
        "Model from Paper",
        "Trained from Scratch"
    ],
    "Inference Time (seconds)": [637.51, 638.33, 636.69, 638.09],
    "Images Processed": [980, 980, 980, 980],
    "Time per Image (ms)": [637.51/980*1000, 638.33/980*1000, 636.69/980*1000, 638.09/980*1000]
}

# Convert to DataFrame
inference_df = pd.DataFrame(inference_data)

# Highlight the fastest inference time
fastest_model = inference_df["Inference Time (seconds)"].idxmin()
second_fastest_model = inference_df["Inference Time (seconds)"].drop(fastest_model).idxmin()

inference_df.loc[fastest_model, "Inference Time (seconds)"] = f"**{inference_df.loc[fastest_model, 'Inference Time (seconds)']}**"
inference_df.loc[second_fastest_model, "Inference Time (seconds)"] = f"_{inference_df.loc[second_fastest_model, 'Inference Time (seconds)']}_"  

# Plotting the table as an image
fig, ax = plt.subplots(figsize=(8, 2))  # Adjust the figure size to fit the table
ax.axis('off')  # Turn off the axis

# Create and display the table
tbl = table(ax, inference_df, loc='center', colWidths=[0.2] * len(inference_df.columns))
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1.2, 1.2)  # Adjust scaling

# Save or show
plt.title("Inference Time Table", fontsize=12, pad=10)
plt.show()
```
